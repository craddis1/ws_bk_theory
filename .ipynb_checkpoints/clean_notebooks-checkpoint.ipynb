```python
import nbformat
from nbformat.v4.nbbase import NotebookNode

def load_notebook(notebook_path):
    with open("Bk_Theory.ipynb", 'r', encoding='utf-8') as f:
        nb = nbformat.read(f, as_version=4)
    return nb

def clear_outputs(notebook):
    for cell in notebook.cells:
        if 'outputs' in cell:
            cell.outputs = []
    return notebook

def save_notebook(notebook, output_path):
    with open(output_path, 'w', encoding='utf-8') as f:
        nbformat.write(notebook, f)

# Load the notebook
notebook = load_notebook('your_notebook.ipynb')

# Clear the outputs
notebook = clear_outputs(notebook)

# Save the modified notebook
save_notebook(notebook, 'cleaned_notebook.ipynb')
```
