```python
from matplotlib.patches import Ellipse

def plot_ellipse(ax, sigma, pos, **kwargs):
    def ellipse_pars2(sigma):
        a2 = (sigma[0,0]+sigma[1,1])/2 + np.sqrt((sigma[0,0]-sigma[1,1])**2/4+sigma[0,1]**2)
        b2 = (sigma[0,0]+sigma[1,1])/2 - np.sqrt((sigma[0,0]-sigma[1,1])**2/4+sigma[0,1]**2)
        theta2 = np.arctan(2*sigma[0,1]/(sigma[0,0]-sigma[1,1]))
        return np.sqrt(a2),np.sqrt(b2),(theta2)/2
    a1,b1,t1 = ellipse_pars2(sigma)
    ell1 = Ellipse(xy=pos, width=2*1.52*b1, height=2*1.52*a1, angle=t1*180/np.pi, edgecolor='b', fc='None', lw=2,**kwargs)
    ax.add_patch(ell1)
    ell2 = Ellipse(xy=pos, width=2*2.48*b1, height=2*2.48*a1, angle=t1*180/np.pi,edgecolor='c', fc='None', lw=2, **kwargs)
    ax.add_patch(ell2)
```
